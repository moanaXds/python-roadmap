```python
import math

class Vector2d:

    def __init__(self, a=0,b=0):
        self.a=a
        self.b=b

    def __add__(self, other):    # operator overload for + 
        return Vector2d(self.a + other.a, self.b+other.b)


    def __sub__(self, other): 
        return Vector2d(self.a - other.a, self.b-other.b)

    def __mul__(self, val):
        return Vector2d(self.a*val ,self.b*val )
    

    def mag(self):
        return math.sqrt(self.a**2 + self.b**2)

    def normal(self):
        magnitude=self.mag()
        if(magnitude!=0):
            return Vector2d(self.a/magnitude, self.b/magnitude)
        else:

            return Vector2d(0,0)

    def dot(self,other ):

        return (self.a * other.a + self.b * other.b)

    def __print__(self):
        return f"Our 2D Vector : {self.a} and {self.b}"




#---------------------------------------------------------




class Particle:

    def __init__(self, position, velocity, mass, area, drag_coefficient):
        self.position=position   # position[0]= x axis
                                 # position[1]= y axis
        
        self.velocity=velocity   # velocity[0]= x
                                 # velocity[1]= y
        
        self.mass=float(mass)
        self.area=float(area)
        self.drag_coefficient=float(drag_coefficient)


    def update(self, gravity, time):
        gravity1 = gravity

        self.velocity[1] = self.velocity[1]- gravity1* time  # velocit decrease along y

        self.position[0] = self.position[0]+ self.velocity[0]*time

        self.position[1] = self.position[1]+ self.velocity[1]*time
        
        return self.position[1]>=0   # y>0  (above ground)
    

#-----------------------------------------------------



class Environment:

    def __init__(self, gravitational_acceleration, air_density):
        self.gravitational_acceleration=gravitational_acceleration
        self.air_density=air_density


#------------------------------------------------------

def SIMULATE ():
    e1= Environment(9.8,1.2)

    p1= Particle([0, 0],[10, 22], 10 ,5.4, e1.air_density)

    #time

    time=0.2
    max_time=10
    t1=0


    trajectory = [] 
    
     # loop til ground or max_time
    while( t1 < max_time and p1.update(e1.gravitational_acceleration , time) ):
        trajectory.append((p1.position[0], p1.position[1]))
        t1 += time

    return trajectory



#-----------------------------------------------------


def plot_trajectory(trajectory):
    width = 60
    height = 20
    grid = []

    for _ in range(height):
        row = []
        for _ in range(width):
            row.append(" ")
        grid.append(row)

    for x, y in trajectory:
        
        xi = int(x)
        yi = int(y)
        
        if(xi >= width):
            xi = width - 1
            
        if(yi >= height):
            yi = height - 1
        grid[height - yi - 1][xi] = "*"

    for row in grid:
        print("".join(row))


traj = SIMULATE() 
plot_trajectory(traj)
```
