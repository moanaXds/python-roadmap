```python
import turtle

t=turtle.Turtle()

t.speed(0)


def drawStar():
    for i in range(5):
        t.forward(50)
        t.right(144)
        
t.left(90)
t.forward(200)

t.left(90)
t.forward(200)

t.left(90)
t.forward(100)

t.left(90)
t.forward(200)

t.backward(70)
t.left(90)
t.forward(100)


t.penup()
t.left(90)
t.forward(50)
t.left(90)
t.forward(20)
t.pendown()

drawStar() 
 
turtle.done()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Quiz")
root.geometry("300x220")


q1 = tk.IntVar(value=-1)
q2 = tk.IntVar(value=-1)
q3 = tk.IntVar(value=-1)

def submit():
    score = 0
    if q1.get() == 1: score += 1
    if q2.get() == 1: score += 1
    if q3.get() == 0: score += 1
    messagebox.showinfo("score", f"{score}/3")

# 1
tk.Label(root, text="1. Capital of pakistan?").pack()
tk.Radiobutton(root, text="Karachi", variable=q1, value=0).pack()
tk.Radiobutton(root, text="Islamabad", variable=q1, value=1).pack()

# 2
tk.Label(root, text="2. Red Planet?").pack()
tk.Radiobutton(root, text="Venus", variable=q2, value=0).pack()
tk.Radiobutton(root, text="Mars", variable=q2, value=1).pack()

# 3
tk.Label(root, text="3. Author of Romeo and Juliet?").pack()
tk.Radiobutton(root, text="Shakespeare", variable=q3, value=0).pack()
tk.Radiobutton(root, text="Dickens", variable=q3, value=1).pack()

# button
tk.Button(root, text="Submit", command=submit).pack(pady=10)

root.mainloop()
```

```python
import csv
import tkinter as tk

employees = [
    ["Ahmad", "HR", 5987, 5343],
    ["Sarah", "IT", 63245, 2344],
    ["Bari", "Finance", 12343, 3242]
]

# Calculate salary 

for emp in employees:
    net = emp[2] + emp[3]
    emp.append(net)

#  highest salary 

highest = max(employees, key=lambda x: x[4])

# CSV 

with open("results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Dept", "Basic", "Bonus", "Net"])
    writer.writerows(employees)
    writer.writerow([])
    writer.writerow(["Highest Paid", highest[0], highest[4]])

# chart 

root = tk.Tk()
c = tk.Canvas(root, width=400, height=300, bg="white")
c.pack()

max_salary = max(emp[4] for emp in employees)
x = 50
for emp in employees:
    height = emp[4] / max_salary * 200
    c.create_rectangle(x, 250 - height, x+40, 250, fill="blue")
    c.create_text(x+20, 260, text=emp[0])
    x += 60

root.mainloop()
```
